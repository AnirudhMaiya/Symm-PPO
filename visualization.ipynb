```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_pickle(pickleFilePath: str):
  results = []
  with open(pickleFilePath, 'rb') as fr:
    try:
        while True:
            results.append(pickle.load(fr))
    except EOFError:
        pass
  return results
```

```python
Vanilla_PPO_Pongv4 = np.array(read_pickle('results/Vanilla_PPO_Pongv4.txt'))
Symm_PPO_Pongv4 = np.array(read_pickle('results/Symm_PPO_Pongv4.txt'))
Vanilla_PPO_Seaquestv0 = np.array(read_pickle('results/Vanilla_PPO_Seaquestv0.txt'))
Symm_PPO_Seaquestv0 = np.array(read_pickle('results/Symm_PPO_Seaquestv0.txt'))
```

```python
median = 5
updates = 1
plt.figure(figsize = (10,8))
plt.plot(Vanilla_PPO_Pongv4[:,updates]/1e6,Vanilla_PPO_Pongv4[:,median], label = 'Vanilla PPO')
plt.plot(Symm_PPO_Pongv4[:,updates]/1e6,Symm_PPO_Pongv4[:,median], label = 'Symm-PPO')
plt.ylabel("Median Rewards")
plt.xlabel("Timesteps (in million)")
plt.title("PongNoFrameskip-v4 with Entropy Coefficient 0.05")
plt.legend()
plt.savefig("PongNoFrameskip-v4.png")
```

```python
median = 5
updates = 1
minTimeSteps = min(Vanilla_PPO_Pongv4.shape[0], Symm_PPO_Pongv4.shape[0])
plt.figure(figsize = (10,10))
diff = Symm_PPO_Pongv4[:,median][0:minTimeSteps] - Vanilla_PPO_Pongv4[:,median][0:minTimeSteps]
plt.plot(Vanilla_PPO_Pongv4[:,updates]/1e6,diff, label = 'Symm-PPO minus Vanilla PPO')
plt.axhline(y=0, color='k', linestyle='-')
plt.ylabel("Difference of Median Rewards")
plt.xlabel("Timesteps (in million)")
plt.title("PongNoFrameskip-v4 with Entropy Coefficient 0.05")
plt.legend()
plt.savefig("PongNoFrameskip-v4-diff.png")
```
